```python
## series creation
import pandas as pd
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print("Pandas Series:\n", series)
```

```python
## create a series from a dictionary
data_dict = {'a': 10, 'b': 20, 'c':             30}
series_dict = pd.Series(data_dict)
print("\nPandas Series from Dictionary:\n", series_dict)
```

```python
## create a DataFrame from a dictionary
data_frame = { 'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['New York', 'Los Angeles', 'Chicago'] }
df = pd.DataFrame(data_frame)
print("\nPandas DataFrame:\n", df)
```

```python
import numpy as np
np.array(df)
```

```python
## create a dataframe from a list of dictionaries
data_list_dicts = [ {'Name': 'Alice', 'Age': 25, 'City': 'New York'}, {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'}, {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'} ]
df_list_dicts = pd.DataFrame(data_list_dicts)
print("\nPandas DataFrame from List of Dictionaries:\n", df_list_dicts)
```
