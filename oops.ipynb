```python
## instance variables and methods in Python
class MyClass:
    def __init__(self, value):
        self.instance_variable = value

    def instance_method(self):
        return f'Instance variable value is: {self.instance_variable}'
obj = MyClass(10)
print(obj.instance_method())
```

```python
## define a class with a class variable and a class method
class MyClass:
    class_variable = "I am a class variable"

    @classmethod
    def class_method(cls):
        return f'Class variable value is: {cls.class_variable}'
print(MyClass.class_method())
```

```python
## define a class using instance variable
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f'Hello, my name is {self.name} and I am {self.age} years old.'

person1 = Person('Alice', 30)
print(person1.introduce())
```

```python
### Modiling a bank account
class BankAccount:
    def __init__(self,owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f'Deposited {amount}. New balance is {self.balance}.')

    def withdraw(self, amount):
        if amount > self.balance:
            print('Insufficient funds.')
        else:
            self.balance = amount
            print(f'Withdrew {amount}. New balance is {self.balance}.')
account = BankAccount('John', 100)
account.deposit(50)
account.withdraw(30)
```
