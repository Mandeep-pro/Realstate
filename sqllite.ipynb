```python
import sqlite3
```

```python
connection = sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT, age INTEGER, email TEXT, Department TEXT)''') 

## commit the changes
connection.commit()

## Insert sample data
cursor.execute("INSERT INTO users (name, age, email, Department) VALUES ('Alice', 30, 'alice@example.com', 'HR')")
cursor.execute("INSERT INTO users (name, age, email, Department) VALUES ('Bob', 25, 'bob@example.com', 'IT')")
cursor.execute("INSERT INTO users (name, age, email, Department) VALUES ('Charlie', 35, 'charlie@example.com', 'Finance')")
cursor.execute("INSERT INTO users (name, age, email, Department) VALUES ('Diana', 28, 'diana@example.com', 'IT')")
cursor.execute("INSERT INTO users (name, age, email, Department) VALUES ('Eve', 32, 'eve@example.com', 'HR')")

connection.commit()
print("Table created and data inserted successfully!")
```

```python
## Display table data
import pandas as pd

cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()

# Get column names
column_names = [description[0] for description in cursor.description]

# Create a DataFrame for better visualization
df = pd.DataFrame(rows, columns=column_names)
print(df)
print("\nTotal records:", len(df))
```
