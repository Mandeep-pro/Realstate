```python
import os
```

```python
## write in a file
with open('destination.txt', 'w') as file:
    file.write('Hello, World!\n')
    file.write('This is a sample file.\n')
```

```python
## wap to append and add lines in the file
with open('destination.txt', 'a') as file:
    file.write('This line is appended.\n')
    file.write('Appending another line.\n')
    
```

```python
## wap to write and read from a file
with open('sample.txt', 'w+') as file:
    file.write('This is a sample text file.\n')
    file.write('It contains multiple lines of text.\n')
    file.seek(0)  # Move the cursor to the beginning of the file
    content = file.read()
    print(content)
```

```python
## wap to create a new directory 
new_directory = 'packages'
os.mkdir(new_directory)
print(f'Directory "{new_directory}" created.')
```

```python
### JOing Paths
dir_name = "folder"
file_name = "file.txt"
full_path = os.path.join(dir_name, file_name)
print(full_path)
```

```python
path = 'example.txt'
if os.path.exists(path):
    print(f'The file "{path}" exists.')
else:
    print(f'The file "{path}" does not exist.')
```

```python
## checking if a path is a file or directory
path = 'destination.txt'
if os.path.isfile(path):
    print(f'"{path}" is a file.')
elif os.path.isdir(path):
    print(f'"{path}" is a directory.')
else:
    print(f'"{path}" does not exist.')
```

```python
## getting absolute path
absolute_path = os.path.abspath('destination.txt')
```
